# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets
from matplotlib.colors import ListedColormap
```

# Data
Load data set from standard and store in X

Class values are stored in y.

```python
iris = datasets.load_iris()
X = iris.data[:, :2]  ## only takes the first two features.
y = iris.target
```

Those coordinates can be visualized with matplotlib.

```python
## Create color maps
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])


%pylab inline
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold)
plt.show()
```

# Support Vector Machine -- linear kernel

```python
clf = svm.SVC(kernel='linear')
clf.fit(X,y)
```

```python
w = clf.coef_[0]
print(w)


a = -w[0] / w[1]

xx = np.linspace(0,12)
yy = a * xx - clf.intercept_[0] / w[1]

## k- stands for a black line, k for the color and the minus for the type
h0 = plt.plot(xx, yy, 'k-')

## x is the first column of X, y is the second column of X, the classes are in y 
plt.scatter(X[:, 0], X[:, 1], c = y)

plt.show()
```

```python
h = .02  ## step size in the mesh

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

## Put the resulting areas into a color plot
Z = Z.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

## Plot also the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())

plt.show()
```

# Support Vector Machine -- RBF kernel

```python
clfr = svm.SVC(kernel='rbf', gamma=0.7)
clfr.fit(X,y)
```

```python
h = .02  ## step size in the mesh

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

Z = clfr.predict(np.c_[xx.ravel(), yy.ravel()])

## Put the resulting areas into a color plot
Z = Z.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

## Plot also the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())

plt.show()
```
