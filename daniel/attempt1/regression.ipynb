```python
# import libraries and adjust settings
from datetime import date
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# set pyplot style
plt.style.use('fivethirtyeight') 

# Render our plots inline
%matplotlib inline

# set figure size
plt.rcParams['figure.figsize'] = (15, 8)

# load datasets
inPath = r'data/'
df_merged = pd.read_csv(inPath+'preprocessed/exp_mergedData.csv')
```

```python
# Create filtered dataframes
df_China = df_merged[df_merged.location == 'China']
df_Germany = df_merged[df_merged.location == 'Germany']
df_India = df_merged[df_merged.location == 'India']
df_Italy = df_merged[df_merged.location == 'Italy']
df_UK = df_merged[df_merged.location == 'United Kingdom']
df_US = df_merged[df_merged.location == 'United States']
```

```python
df_China.plot(x='GDP',y='Monthly_total_cases',style='-')
plt.title('GDP vs Cases')
plt.xlabel('GDP')
plt.ylabel('New cases per month')
plt.show()
df_China
```
